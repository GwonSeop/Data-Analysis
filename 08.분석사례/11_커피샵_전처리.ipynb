```python
from glob import glob
df_all = pd.DataFrame(columns=['x','y'])
for file in glob(f'{data_folder}/*.csv'):
  df = pd.read_csv(file)
  df = df[['상호명','지점명','시도명','시군구명','도로명주소']]
  df_star = df[df['상호명'].str.contains('starbucks', na=False, case=False)]
  df_all = df_all.append(df_star)
  df_star = df[df['상호명'].str.contains('스타벅스', na=False, case=False)]
  df_all = df_all.append(df_star)
  df_edia = df[df['상호명'].str.contains('이디야', na=False)]
  df_all = df_all.append(df_edia)
  df_edia = df[df['상호명'].str.contains('ediya', na=False, case=False)]
  df_all = df_all.append(df_edia)
  df_bbak = df[df['상호명'].str.contains('paik', na=False, case=False)]
  df_all = df_all.append(df_bbak)
  df_bbak = df[df['상호명'].str.contains('빽다방', na=False, case=False)]
  df_all = df_all.append(df_bbak)
  df_bin = df[df['상호명'].str.contains('커피빈', na=False, case=False)]
  df_all = df_all.append(df_bin)
  df_bin = df[df['상호명'].str.contains('coffeebean', na=False, case=False)]
  df_all = df_all.append(df_bin)
del df_all['x']
del df_all['y']
df_all
```
