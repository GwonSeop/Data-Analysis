```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
# %matplotlib inline 주피터 노트북에서그래프를 볼 떄 사용하는 매직 명령어
```

```python
plt.title("라인 플롯")
plt.plot([1, 4, 9, 16])      # y = (x + 1) ** 2
plt.grid()    # 격자선
plt.show()
```

```python
plt.title("라인 플롯")
xs = np.linspace(0, 3, 301)
ys = (xs + 1) ** 2
plt.plot(xs,ys)      # y = (x + 1) ** 2
plt.grid()    # 격자선
plt.show()
```

## 스타일 지정

```python
# 색깔 (r), 마커 (s) : 점 , 선 스타일(--)  >>  matplotlib.org 참고
plt.title("'rs--' 스타일의 plot ")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'ks--')
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("스타일 적용 예")
plt.grid()
plt.show()
```

### 그림 범위 지정

```python
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50)      # x축 범위
plt.ylim(-2, 20)    # y축 범위
plt.grid()
plt.show()
```

###틱설정

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi]) # 그래프에 찍히는 점 표시
plt.yticks([-1, 0, +1])
plt.grid()
plt.show()
```

```python
# 수학 문자식
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("LaTeX, 문자열로 tick label 정의")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.show()
```

### 여러개의 선 그리기

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.show()
```

```python
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--')
plt.plot(t, 0.5 * t**2, 'bs:')
plt.plot(t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

### 범례(Legend)

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")    # 범례사용시 label 입력이 필요하다
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

### x축, y축 라벨, 타이틀

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.grid()
plt.show()
```
