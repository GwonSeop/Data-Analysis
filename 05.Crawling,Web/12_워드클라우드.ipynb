```python
s = 'A quick brown fox jumps over the lazy dog.'
Counter(s)
```

- 해시태그 빈도 집계

```python
tag_counts = Counter(tags_total)
```

```python
# Top 50
tag_counts.most_common(50)
```

```python
# 데이터 정제하기
stopwords = ['#jeju','#일상','','#제주','#제주도','#선팔','#제주도눈썹문신',
             '#서귀포눈썹문신','#반영구','#눈썹문신','#제주자연눈썹']
'''tag_total_refined = []
for tag in tags_total:
  if tag not in stopwords:
    tag_total_refined.append(tag)'''
tags_total = [tag for tag in tags_total if tag not in stopwords]
tag_counts = Counter(tags_total)
tag_counts.most_common(50)
```

# 워드 클라우드 만들기

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
! ls -l /usr/share/fonts/truetype/nanum
```

```python
path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
wordcloud = WordCloud(background_color='white',
                      max_words=100,
                      relative_scaling=0.3,
                      width=800, height=400).generate_from_frequencies(tag_counts)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
